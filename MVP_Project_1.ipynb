# NYC Covid Vaccination Campaign 

**OBJECTIVE**:  An initial proposal was made to use the NYC subway system as a distribution mechanism for the Covid-19 Vaccine.  Ideal sites would potentially be chosen according to turnstile entry foot traffic, time of day and neighborhood vaccination rates.   

**TIME FRAME**:  A prelimary analysis of NYC MTA subway turnstile data was conducted for the most recent three months.  This time range was specifically chosen to reflect the time period in which the vaccine was made available to the general population of NYC.  Rather than assume a return to pre-Covid normalcy using pre-Covid data, it was assumed commuting activity forecast for the next few months would be more precise if extrapolated from the most recent data.  

**SOLUTION**: The initial analysis was comprised of finding the top 10 busiest stations as measured by the total turnstiles entries for the time-frame previously mentioned. *The box plot below displays the initial recommendations in order.*
